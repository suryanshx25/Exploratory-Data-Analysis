```python
df['family_size']=df['Parch'] + df['SibSp']
```

```python
df.sample(5)
```

```python
def family_type(number):
    if number == 0:
        return "Alone"
    elif number > 0 and number <= 4:
        return "Medium"
    else:
        return "Large"
```

```python
df['family_type'] = df['family_size'].apply(family_type)
```

```python
df.sample(5)
```

```python
df.drop(columns = ['SibSp','Parch','family_size'],inplace = True)
```
